```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
f = cv.imread('./a1images/emma.jpg',cv.IMREAD_GRAYSCALE)
assert f is not None

t1 = np.linspace(0,50,50+1-0,dtype=np.uint8)
t2 = np.linspace(50+1,100,50-50,dtype=np.uint8)
t3 = np.linspace(100+1,255,150-50,dtype=np.uint8)
t4 = np.linspace(255,150,150-150,dtype=np.uint8)
t5 = np.linspace(150+1,255,255-150,dtype=np.uint8)
t = np.concatenate((t1,t2),axis=0).astype(np.uint8)
t = np.concatenate((t,t3),axis=0).astype(np.uint8)
t = np.concatenate((t,t4),axis=0).astype(np.uint8)
t = np.concatenate((t,t5),axis=0).astype(np.uint8)


g=cv.LUT(f,t)

fig,ax = plt.subplots(1,3,figsize=(15,5))
ax[0].plot(t,'r',label="Transformation")
ax[1].imshow(f,cmap='gray')
ax[1].set_title("Original")
ax[2].imshow(g,cmap='gray')
ax[2].set_title("Transformed")
for a in ax[1:]:
  a.axis('off')
plt.show()

print(c[0,0])
```
